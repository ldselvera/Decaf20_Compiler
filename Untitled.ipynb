```python
def scanner(filename):
    current_state  = 0
    ch = ''
    ident = ""
    token = ""
    myfile = open("out.txt", "w")
    line_count = 0
    col_count = 0
    
    with open(filename) as f:
      while True:
        ch = f.read(1)
        
        if not ch: return 0
        
        if ch == '\n':
            line_count += 1
        #Save previous state for backspace    
        previous_state = current_state 

        print("before:", ch, current_state)
        
        if (ch == 'e' or ch == 'E') and (previous_state == 6 or previous_state == 4):
            current = df[ch][current_state]
        elif ch.isalpha(): 
            current_state  = df['a'][current_state ]
        elif ch.isdigit():
            current_state  = df['0'][current_state ]
        else:
            current_state  = df[ch][current_state ]

        print("after:", current_state)
            
        print("string:", ident)
        #If have reached the end of current FSM
        if current_state  == 0:
            if ident:
                token = df["Token"][previous_state]
                
                if token is 'T_Identifier':
                    if ident in reserved :
                        token = reserved[ident]
                if token in operators:
                    token = "'" + token + "'"

                if token == 'T_IntConstant (value = %d)':    
                    token = token % (int(ident)) 
                if token == "T_DoubleConstant (value = %f)":
                    token = token % (float(ident)) 
                line =  ' line %s cols %s-%s is %s\n' % ( str(line_count), str(1), str(col_count), token)

                myfile.write(f"{ident:<12}{line:>10}")
                
                #Reset current string and column count
                ident = ''
                col_count = 0
        else:
            #Keep adding to current string
            ident += ch
            col_count += 1
            
    myfile.close()
```
